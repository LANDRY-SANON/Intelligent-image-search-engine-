```python
from drive.MyDrive.IISEngine_test_data.Engine_utils import *
from google.colab.patches import cv2_imshow
from scipy.cluster.vq import *
from ipywidgets import GridspecLayout, widgets, Output, Image
from IPython.display import display
from skimage import io

# Define the first Block function and widgets

query_image_path_widget_1 = widgets.Text(description='Query Image Path:', layout=widgets.Layout(width='400px'))
database_dir_widget_1 = widgets.Text(description='Database Directory:', layout=widgets.Layout(width='400px'))
find_similar_images_histogram_button = widgets.Button(description='Find Similar Images using histogram', layout=widgets.Layout(width='300px'))
find_similar_images_PTM_button = widgets.Button(description='Find Similar Images using PT model', layout=widgets.Layout(width='300px'))
search_similar_image_by_SIFT_button = widgets.Button(description='Find Similar Images using ML and SIFT ', layout=widgets.Layout(width='300px'))
search_by_image_fragments_button= widgets.Button(description='Detect images using fragment', layout=widgets.Layout(width='300px'))
recognize_faces_in_images_button= widgets.Button(description='Recognize faces in images', layout=widgets.Layout(width='300px'))

output_widget_1 = Output()


def on_find_similar_images_histogram_clicked_1(b):
    with output_widget_1:
        output_widget_1.clear_output()


        search_similar_image_by_color_histogram(query_image_path_widget_1.value, database_dir_widget_1.value)


def on_find_similar_images_PTM_clicked_1(b):
    with output_widget_1:
        output_widget_1.clear_output()
        similar_image_paths = search_similar_image_by_Pretrained_Model(query_image_path_widget_1.value, database_dir_widget_1.value)

def on_detect_images_using_fragment_clicked_1(b):
    with output_widget_1:
        output_widget_1.clear_output()
        similar_image_paths = search_for_image_fragments(query_image_path_widget_1.value, database_dir_widget_1.value)

def on_recognize_faces_in_images_clicked_1(b):
    with output_widget_1:
        output_widget_1.clear_output()
        recognize_faces_in_images(query_image_path_widget_1.value, database_dir_widget_1.value)

def on_search_similar_image_by_SIFT_clicked_1(b):
    with output_widget_1:
        output_widget_1.clear_output()
        #search_similar_image_by_SIFT(query_image_path_widget_1.value, database_dir_widget_1.value)
        print( " Sorry ! the integration of this feature are not done yet ")



find_similar_images_histogram_button.on_click(on_find_similar_images_histogram_clicked_1)
find_similar_images_PTM_button.on_click(on_find_similar_images_PTM_clicked_1)
search_similar_image_by_SIFT_button.on_click(on_search_similar_image_by_SIFT_clicked_1)
search_by_image_fragments_button.on_click(on_detect_images_using_fragment_clicked_1)
recognize_faces_in_images_button.on_click(on_recognize_faces_in_images_clicked_1)

# Define the second Block function and widgets


query_image_widget_2 = widgets.Text(description='Query image:', layout=widgets.Layout(width='400px'))
recommend_footwear_button_2 = widgets.Button(description='Detect footwear')
output_widget_2 = Output()

def on_recommend_footwear_clicked_2(b):
    with output_widget_2:
        output_widget_2.clear_output()
        test_img = io.imread(query_image_widget_2.value)
        recommend_footwear_of_the_target_brand(test_img)



recommend_footwear_button_2.on_click(on_recommend_footwear_clicked_2)

# Define the third Block function and widgets


with open("/content/drive/MyDrive/IISEngine_test_data/Models/faster_rcnn_inception_v2_coco_2018_01_28/coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
label_widget = widgets.Dropdown(options=classes,description='Label:',disabled=False)
database_dir_widget_3 = widgets.Text(description='Database Director:', layout=widgets.Layout(width='400px'))
detect_objects_using_FasterRCNN_button_3 = widgets.Button(description='Validate')
output_widget_3 = Output()

def on_detect_objects_using_FasterRCNN_clicked_3(b):
    with output_widget_3:
        output_widget_3.clear_output()

        images_list = [filename for filename in os.listdir(database_dir_widget_3.value) if filename.lower().endswith(".jpg") or filename.lower().endswith(".png")]
        label = label_widget.value
        for filename in images_list:
            image_path = os.path.join(database_dir_widget_3.value, filename)
            image = cv2.imread(image_path)
            image = cv2.resize(image, None, fx=0.4, fy=0.4)
            detected_image, detected_labels, detected_boxes = detect_objects_using_FasterRCNN(image, classes)

            if label in detected_labels:
                for box in detected_boxes:
                    x1, y1, x2, y2 = box
                    cv2.rectangle(detected_image, (x1, y1), (x2, y2), (0, 255, 0), 2)

                cv2_imshow(detected_image)
                cv2.waitKey(0)
                cv2.destroyAllWindows()



detect_objects_using_FasterRCNN_button_3.on_click(on_detect_objects_using_FasterRCNN_clicked_3)

# Define the fourth block (OCR function and widget )

query_text_widget_4 = widgets.Text(description='Query Text:', layout=widgets.Layout(width='400px'))
database_dir_widget_4 = widgets.Text(description='Database Director:', layout=widgets.Layout(width='400px'))
search_images_contaning_text_button_4 = widgets.Button(description='Find Images that contain this word', layout=widgets.Layout(width='300px'))
output_widget_4 = Output()

def on_search_images_contaning_text_clicked_4(b):
    with output_widget_4:
        output_widget_4.clear_output()
        search_images_containing_text(query_text_widget_4.value, database_dir_widget_4.value)



search_images_contaning_text_button_4.on_click(on_search_images_contaning_text_clicked_4)


grid_layout = GridspecLayout(4, 3, height='1000px' )

grid_layout[0, 0] = widgets.VBox([query_image_path_widget_1, database_dir_widget_1, find_similar_images_histogram_button, find_similar_images_PTM_button,search_similar_image_by_SIFT_button,search_by_image_fragments_button,recognize_faces_in_images_button])
grid_layout[0, 1] = widgets.VBox([output_widget_1])


grid_layout[1, 0] =  widgets.VBox([ query_image_widget_2, recommend_footwear_button_2])
grid_layout[1, 1] = widgets.VBox([output_widget_2])
grid_layout[1,0].layout.margin = '50px 0px 0px 0px'

grid_layout[2, 0] = widgets.VBox([label_widget, database_dir_widget_3, detect_objects_using_FasterRCNN_button_3])
grid_layout[2, 1] = widgets.VBox([output_widget_3])
grid_layout[2,0].layout.margin = '50px 0px 0px 0px'

grid_layout[3, 0] = widgets.VBox([query_text_widget_4, database_dir_widget_4, search_images_contaning_text_button_4])
grid_layout[3, 1] = widgets.VBox([ output_widget_4])
grid_layout[3,0].layout.margin = '50px 0px 0px 0px'



# Display the interface
grid_layout
```
